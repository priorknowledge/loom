# Fetch Data
Download the Iris data from the UCI ML repository and store locally

```python
TEMP_DIR = '/tmp'
DATA_URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
RAW = TEMP_DIR + '/iris_raw.csv'
SCHEMA = TEMP_DIR + '/iris_schema.json'
#urllib.urlretrieve(DATA_URL, RAW)
```

```python
schema = {
    'sepal_length': 'nich',
    'sepal_width': 'nich',
    'petal_length': 'nich',
    'petal_width': 'nich',
    'class': 'dd'
}
with open(SCHEMA, 'wb') as fd:
    fd.write(json.dumps(schema, indent=2))
```

```python
name = 'iris'
# loom.datasets.load(name, SCHEMA, RAW)
loom.datasets.clean(name)
loom.tasks.ingest(name, SCHEMA, RAW)
loom.tasks.infer(name, sample_count=1)
```
